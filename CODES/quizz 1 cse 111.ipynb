```python
#ans to the question:



ID=['20201199','21121347','19101052','19301121','20221369','21241199','19241187','20141007','20101035', '21121875', '19141534', '20341552', '20121135', '21341001']

x={"CSE":{'2019':[],'2020':[],'2021':[]},"EEE":{'2019':[],'2020':[],'2021':[]},"CS":{'2019':[],'2020':[],'2021':[]}}

for i in ID:
  dept=i[3:5]
  yearcode= int(i[0:2])
  if dept=="01":
    if yearcode==19:
      x["CSE"]['2019'].append(i)
    elif yearcode==20:
      x["CSE"]['2020'].append(i)
    else:
      x["CSE"]['2021'].append(i)
  elif dept=="21":
    if yearcode==19:
      x["EEE"]['2019'].append(i)
    elif yearcode==20:
      x["EEE"]['2020'].append(i)
    else:
      x["EEE"]['2021'].append(i)
  else:
    if yearcode==19:
      x["CS"]['2019'].append(i)
    elif yearcode==20:
      x["CS"]['2020'].append(i)
    else:
      x["CS"]['2021'].append(i)
    
  
  
print(x)
```
