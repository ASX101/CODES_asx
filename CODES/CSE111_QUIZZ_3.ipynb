```python
class BFC:
  menu ={}
  order_id = 80
  def __init__(self):
    pass
  def addChicken(self,*args):
    self.args = args
    for i in range(len(self.args)):
      if i%2 == 0:
        key = self.args[i]
      else:
        value = self.args[i]
        BFC.menu[key] = value
  def placeOrderof(self,order):
    BFC.order_id +=1
    print(f"Order ID:{BFC.order_id}")
    print(f"Customer name: {order.name}")
    print("Order:")
    if len(order.order)==1:
      print(f"Burger:{order.order[2:0]} , Quantity:{order.order[:1]}")
    else:
      s = []
      s.append(order.order)
      for i in s:
        food = i[2:]
        quan = i[0]
      print(f"Burger: {food} , Quantity: {quan}")



class Customer:
  def __init__(self,name):
    self.name = name
    self.order = ""
  def setOrder(self,*args):
    self.order = args


print(f"It's been a wonderful day! \nToday, we have already  served {BFC.order_id} customers. Few more to go before we  close for today.")
print("**************************************************")
outlet1 = BFC()
print("1.==========================================")
print("Chicken Menu:")
print(BFC.menu)
print("2.==========================================")
outlet1.addChicken('Regular',95,'Spicy',125)
print("3.==========================================")
print("Chicken Menu:")
print(BFC.menu)
print("4.==========================================")
c1 = Customer("Ariyan")
c1.order = "2xRegular"
outlet1.placeOrderof(c1)
print("5.==========================================")
c2 = Customer("Ayan")
c2.setOrder("2xRegular,3xSpicy")
print("6.==========================================")
outlet1.placeOrderof(c2)
```
