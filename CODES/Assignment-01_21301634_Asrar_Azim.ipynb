```python
#task 1 :

word = input()
upper_word = 0
for i in word:
    if ord('A') <= ord(i) <= ord('Z'):  
        upper_word += 1
lower_word = len(word) - upper_word

if upper_word > lower_word:
    print(word.upper())
else:
    print(word.lower())
```

```python
#task 2 :

input1 = input()

number = 0
word = 0
length = len(input1)

for i in input1:
    if ord('0') <= ord(i) <= ord('9'): 
        number += 1
    if ord('A') <= ord(i) <= ord('Z') or ord('a') <= ord(i) <= ord('z'):  
        word += 1

if number == length:
    print("NUMBER")  
elif word == length:
    print("WORD")  
else:
    print("MIXED") 
```

```python
#task 6 :

password = input()

upper = 0
lower = 0
number = 0
special = 0



for i in password:
    if ord("A") <= ord(i) <= ord("Z"):  
        upper += 1
    if ord("a") <= ord(i) <= ord("z"):  
        lower += 1
    if ord("0") <= ord(i) <= ord("9"):  
        number += 1


if " " in password:
    special += 1
if "$" in password:
    special += 1
if "#" in password:
    special += 1
if "@" in password:
    special += 1



if upper == 0:
    fault = "Uppercase character missing"
    print(fault)
if lower == 0:
    fault = "Lowercase character missing"
    print(fault)
if number == 0:
    fault = "Digit is missing"
    print(fault)
if special == 0:
    fault = "Special characters missing"
    print(fault)
elif (upper and lower and number and special) != 0:  
    print("OK")
```

```python
#task 5 :

upper_letter = []
lower_letter = []
digit = []

y = input()

for i in y:
    if ord("a") <= ord(i) <= ord("z"):
        lower_letter.append(i)
    elif ord("A") <= ord(i) <= ord("Z"):
        upper_letter.append(i)
    else:
        digit.append(i)

int_digit = []
even_list = []
odd_list = []

for k in digit:
    int_digit.append(int(k))

for g in int_digit:
    if g % 2 == 0:
        even_list.append(g)
    else:
        odd_list.append(g)

even_list_letter = []
odd_list_letter = []

for even in sorted(even_list):
    even_list_letter.append(str(even))


for odd in sorted(odd_list):
    odd_list_letter.append(str(odd))


final_list = sorted(lower_letter) + sorted(upper_letter) + sorted(odd_list_letter) + sorted(even_list_letter)

final = ""
for j in final_list:
    final += j
print(final)
```

```python
#task 1 :


list_01 = input().split(", ")
list_02 = input().split(", ")

dict_01 = {}
dict_02 = {}
for i in range(len(list_01)):
    x = list_01[i].split(": ")
    key = x[0]
    value = x[1]
    dict_01[key] = int(value)

for j in range(len(list_02)):
    x = list_02[j].split(": ")
    key = x[0]
    value = x[1]
    dict_02[key] = int(value)

final_dict = {}
for k in dict_01.keys():
    for l in dict_02.keys():
        if k == l:
            final_dict[k] = dict_01[k] + dict_02[l]
        elif k not in final_dict.keys():
            final_dict[k] = dict_01[k]
        elif l not in final_dict.keys():
            final_dict[l] = dict_02[l]
x = sorted(final_dict.items())
print(dict(x))

unique_list = []
for i in final_dict.keys():
    if final_dict[i] not in unique_list:
        unique_list.append(final_dict[i])

unique_tuple = tuple(sorted(unique_list))
print(unique_tuple)
```

```python
#task 4 :


str_01 = input()
str_02 = input()

list_01 = list(str_01)
list_02 = list(str_02)


def bub(list_01):
    index_len = len(list_01)-1
    sor = False
    while not sor:
        sor = True
        for i in range(0, index_len):
            if list_01[i] > list_01[i+1]:
                sor = False
                list_01[i], list_01[i+1] = list_01[i+1], list_01[i]
    return list_01


bub(list_01)


def bub(list_02):
    index_len = len(list_02)-1
    sor = False
    while not sor:
        sor = True
        for i in range(0, index_len):
            if list_02[i] > list_02[i+1]:
                sor = False
                list_02[i], list_02[i+1] = list_02[i+1], list_02[i]
    return list_02


bub(list_02)

if list_01 == list_02:
    print("Those strings are anagrams.")
```
